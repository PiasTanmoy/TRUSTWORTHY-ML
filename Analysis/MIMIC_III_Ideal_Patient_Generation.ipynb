```python
ideal_case_df = pd.read_csv('ideal_case_0.01334.csv')

# Normal
normal_sbp = np.arange(100, 121, 5).tolist()
normal_dbp = np.arange(70, 81, 5).tolist()
normal_glucose = np.arange(70, 100, 10).tolist()
normal_resp = np.arange(12, 20, 5).tolist()
pH2 = np.arange(7.35, 7.45, 0.05).tolist()
normal_pH = [round(x, 2) for x in pH2]
normal_Heart_rate = np.arange(50, 70, 10).tolist()
normal_oxy_sat = np.arange(96, 100, 2).tolist()

#MAP = (SBP + 2*DBP)/3

d = {
    'name': []
}

count= 0

for sbp in normal_sbp:
  for dbp in normal_dbp:
    for glucose in normal_glucose:
      for resp in normal_resp:
        for pH in normal_pH:
          for heart_rate in normal_Heart_rate:
            for oxy_sat in normal_oxy_sat:
              MAP = (sbp + 2*dbp)/3

              most_ideal_case = ideal_case_df.copy()
              most_ideal_case['Systolic blood pressure'] = sbp
              most_ideal_case['Diastolic blood pressure'] = dbp
              most_ideal_case['Glucose'] = glucose
              most_ideal_case['Respiratory rate'] = resp
              most_ideal_case['pH'] = pH
              most_ideal_case['Heart Rate'] = heart_rate
              most_ideal_case['Oxygen saturation'] = oxy_sat
              most_ideal_case['Mean blood pressure'] = MAP


              name = 'ideal_case'+'_sbp_'+ str(sbp) +'_dbp_'+ str(dbp) + '_glu_' +str(glucose) + '_resp_' +str(resp)+ '_heart_rate' + str(heart_rate) + '_oxy_' + str(oxy_sat) + '_map_' + str(MAP) +'_pH_' + str(pH) + '.csv, 0'
              most_ideal_case.to_csv('ideal_cases2/ideal_case'+'_sbp_'+ str(sbp) +'_dbp_'+ str(dbp) + '_glu_' +str(glucose) + '_resp_' +str(resp)+ '_heart_rate' + str(heart_rate) + '_oxy_' + str(oxy_sat) + '_map_' + str(MAP) +'_pH_' + str(pH) + '.csv', index=False)
              print(count,name)
              d['name'].append( name)
              count+=1


df = pd.DataFrame(d)
# saving the dataframe
df.to_csv('ideal_cases2/name_list.csv')
```

```python
import os
import numpy as np
import pandas as pd

pathName = os.getcwd()


numFiles = []
fileNames = os.listdir(pathName)

count = 1
for fileNames in fileNames:
    if fileNames.endswith(".csv"):
        numFiles.append(fileNames)
        print(count, fileNames)
        count+=1

d = {
    'name': []
}


import random
random.sample(range(0, 1000), 10)

#ideal_case_df = pd.read_csv('ideal_case_0.01334.csv')

# Normal
high_sbp = random.sample(range(200, 300), 50)
high_dbp = random.sample(range(120, 150), 50)
high_glucose = random.sample(range(250, 350), 50)
high_resp = random.sample(range(30, 50), 10) * 5
high_Heart_rate = random.sample(range(120, 150), 50)
low_oxy_sat = random.sample(range(60, 80), 10) *5
high_body_temp = random.sample(range(38, 44), 5) * 10

#MAP = (SBP + 2*DBP)/3

d = {
    'name': []
}

count= 0

for i in range (len(high_sbp)):
  for fileName in numFiles:
    file = open(os.path.join(pathName, fileName), "rU")
    ideal_case_df = pd.read_csv(file)
    most_ideal_case = ideal_case_df.copy()

    MAP = (high_sbp[i] + 2*high_dbp[i])/3

    most_ideal_case['Systolic blood pressure'] = high_sbp[i]
    most_ideal_case['Diastolic blood pressure'] = high_dbp[i]
    most_ideal_case['Glucose'] = high_glucose[i]
    most_ideal_case['Respiratory rate'] = high_resp[i]
    most_ideal_case['Heart Rate'] = high_Heart_rate[i]
    most_ideal_case['Oxygen saturation'] = low_oxy_sat[i]
    most_ideal_case['Mean blood pressure'] = MAP


    name = 'ideal_case'+'_sbp_'+ str(high_sbp[i]) +'_dbp_'+ str(high_dbp[i]) + '_glu_' +str(high_glucose[i]) + '_resp_' +str(high_resp[i])+ '_heart_rate' + str(high_Heart_rate[i]) + '_oxy_' + str(low_oxy_sat[i]) + '_map_' + str(MAP) + '.csv, 0'
    most_ideal_case.to_csv('../High/high'+'_sbp_'+ str(high_sbp[i]) +'_dbp_'+ str(high_dbp[i]) + '_glu_' +str(high_glucose[i]) + '_resp_' +str(high_resp[i])+ '_heart_rate' + str(high_Heart_rate[i]) + '_oxy_' + str(low_oxy_sat[i]) + '_map_' + str(MAP) +'.csv', index=False)
    print(count,name)
    d['name'].append( name)
    count+=1


df = pd.DataFrame(d)
# saving the dataframe
df.to_csv('../name_list_high.csv', index=False)
```

```python
import os
import numpy as np
import pandas as pd
import random


train_file_C1 = pd.read_csv('train_C1.csv', sep = ',', dtype=object)
train_file_C1_cases = train_file_C1['file_name'].to_list()
print(len(train_file_C1_cases))


numFiles = []

train_file_path = pathName+'/train'
fileNames = os.listdir(train_file_path)

count = 1
for fileName in fileNames:
  if fileName.endswith(".csv") and fileName in train_file_C1_cases:
    numFiles.append(fileName)
    print(count, fileName)
    count+=1

print(len(numFiles))

DBP = np.arange(5, 25, 1).tolist()

d = {
    'name': []
}


for fileName in numFiles[:10]:

  f = open(os.path.join(train_file_path, fileName), "rU")
  ideal_case_df = pd.read_csv(f)
  case = ideal_case_df.copy()

  DBP_val = random.choice(DBP)

  case['Diastolic blood pressure'] = float(DBP_val)

  name = fileName[:-4] + '_DBP_' + str(DBP_val) + '.csv'
  case.to_csv(pathName+'/DBP_C1/' + name, index=False)
  print("SAEVD", name)
  d['name'].append( name + ', 1')


df = pd.DataFrame(d)
# saving the dataframe
df.to_csv(pathName+'/DBP_C1_name_list.csv', index=False)

print("Hey")
```
