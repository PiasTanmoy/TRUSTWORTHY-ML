```python
s1 = 37
s2 = 38
out_file_name = 'Temp-base.csv'

import pandas as pd
from google.colab import files
import os

heat_map = pd.read_csv('in.csv')
display(heat_map)

low = heat_map[heat_map['Unnamed: 0'] < s1]
display(low)

normal = heat_map[(heat_map['Unnamed: 0'] >= s1) & (heat_map['Unnamed: 0'] <= s2)]
display(normal)

high = heat_map[heat_map['Unnamed: 0'] > s2]
display(high)

from numpy.lib.function_base import average
a = []
b = []
c = []
for i in range(16):
  a.append(round(abs(average(low[str(i)]) - average(normal[str(i)])), 2))
  b.append(round(abs(average(normal[str(i)]) - average(high[str(i)])), 2))
  c.append(round(abs(average(low[str(i)]) - average(high[str(i)])), 2))

max_ = [round(max(a), 2), round(max(b),2), round(max(c), 2)]
min_ = [round(min(a), 2), round(min(b), 2), round(min(c), 2)]
avg_ = [round(average(a), 2), round(average(b), 2), round(average(c), 2)]

max_idx = [a.index(max_[0]), b.index(max_[1]), c.index(max_[2])]
min_idx = [a.index(min_[0]), b.index(min_[1]), c.index(min_[2])]

d = {
    'low-norm': a,
    'norm-high': b,
    'low-high': c
}

df = pd.DataFrame(d)

df.loc['max'] = max_
df.loc['max-index'] = max_idx
df.loc['min'] = min_
df.loc['min-index'] = min_idx
df.loc['avg'] = avg_

df.to_csv(out_file_name)
files.download(out_file_name)

# if os.path.exists('in.csv'):
#     os.remove('in.csv')

display(df)
```
