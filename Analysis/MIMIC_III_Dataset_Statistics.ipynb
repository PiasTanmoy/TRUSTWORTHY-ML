```python
counter = 0

for fileName in fileNames:
  if fileName.endswith(".csv") and fileName != "listfile.csv":
    x = pd.read_csv(fileName)

    x['Glascow coma scale eye opening'] = x['Glascow coma scale eye opening'].map(GCS_eye)
    x['Glascow coma scale motor response'] = x['Glascow coma scale motor response'].map(GCS_motor)
    x['Glascow coma scale verbal response'] = x['Glascow coma scale verbal response'].map(GCS_speech)


    mean = x.mean().to_dict()
    std = x.std().to_dict()

    print(fileName, counter)
    counter+=1

    for k in mean.keys():
      if k in means:
        means[k].append(mean[k])
        STDs[k].append(std[k])
```
