# Base LSTM1


```python
base_lstm = pd.read_csv("SHAP-Base LSTM1.csv")
base_lstm

l2 = base_lstm['SHAP'].to_list()
base_lstm
```

### SUM

```python
SHAP_sum = []
for i in range(0, 17):
  vital = vitals[i]
  sum_ = 0
  for j in range(len(l2)):
    if vital in l1[j]:
      sum_ += float(l2[j])

  SHAP_sum.append(sum_)
SHAP_sum

vitals_SHAP_sum = {'vitals': vitals, 'SHAP_sum': SHAP_sum}
df_sum = pd.DataFrame(vitals_SHAP_sum)
df_sum.to_csv('SUM_Base_LSTM1.csv', index= False)
df_sum
```

### AVG

```python
SHAP_avg = []
for i in range(0, 17):
  vital = vitals[i]
  sum_ = 0
  c = 0
  for j in range(len(l2)):
    if vital in l1[j]:
      sum_ += float(l2[j])
      c += 1

  SHAP_avg.append(sum_/c)
SHAP_avg

vitals_SHAP_avg = {'vitals': vitals, 'SHAP_avg': SHAP_avg}
df_avg = pd.DataFrame(vitals_SHAP_avg)
df_avg.to_csv('AVG_Base_LSTM1.csv', index= False)
df_avg
```
