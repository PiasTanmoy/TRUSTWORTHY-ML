```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

f1 = 'EOD 10 - positive lymph nodes examined continuous'
f2 = 'EOD 10 - number of lymph nodes examined continuous'
f3 = 'CS Tumor size continuous'

def inverse_stard(df):
    df[f1] = df[f1]*3.281532853931098 + 1.206322788652106 
    df[f2] = df[f2]* 7.364361918247256 + 6.743024148646638
    df[f3] = df[f3]*22.95831813662134 + 21.392468773994878 
    
    return df
```

### Pos Lymph

```python
# seer_data/Test2/Pos_lymph/pos_lymph_valid_norm_range_0_25_seed_0_x.csv
test_set = pd.read_csv('../Test2/Pos_lymph/pos_lymph_valid_norm_range_0_25_seed_0_x.csv')

df = test_set[[f1, f2, f3]].copy()
display(df)
df = inverse_stard(df)
display(df)

df.to_csv('../Test2/Pos_lymph/f3_seed_0_x.csv')
```

### Grades

```python
grades = ['Grade 1', 'Grade 2', 'Grade 3', 'Grade 4', 'Grade 9']

d = {}


for g in grades:
    result = pd.read_csv('../Test_results/Grades/'+g+'/MLP-5/valid_results_data_frame_4_epo.csv')
    y = result['score y'].to_list()
    d[g] = y

val_y_true = np.load('../processed_data_TS_2/y_valid_normalized.npy')
d['true y'] = val_y_true
df = pd.DataFrame(d) 
df
```

```python
df0  = df[df['true y'] == 0]
df0.shape[0], df0.shape[0]*4
```

```python
print("Grade # ", "C0", " C0%  ", "C1 ", "C1%")
for g in grades:
    c0 = df0[df0[g] < 0.71].shape[0]
    c1 = df0[df0[g] >= 0.71].shape[0]
    c01 = df0.shape[0]
    percent_c0 = round((c0/c01 * 100), 2)
    percent_c1 = round((c1/c01 * 100), 2)
    print(g, c0, percent_c0, c1, percent_c1)
```

```python
df1  = df[df['true y'] == 1]
df1
```

```python
print("Grade # ", "C0", "C0 %  ", "C1   ", "C1 %")
for g in grades:
    c0 = df1[df1[g] < 0.71].shape[0]
    c1 = df1[df1[g] >= 0.71].shape[0]
    c01 = df1.shape[0]
    percent_c0 = round((c0/c01 * 100), 2)
    percent_c1 = round((c1/c01 * 100), 2)
    print(g, c0, percent_c0, c1, percent_c1)
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=400)
plt.rcParams.update({'font.size': 22})

df4 = df[['Grade 1', 'Grade 2', 'Grade 3', 'Grade 4', 'true y']]
custom_palette = ["#fdcdac", "#b3e2cd"] 
custom_palette = ["#ffd7b0", "#b5e5b5"]  # got from origin pro color palet

p = sns.boxplot(
    x = "Grades",
    y = "Survivability",
    showmeans=True,
    hue="true y",
    hue_order=[0, 1],
    showfliers=False,
    palette=custom_palette,
    data=df4.melt(id_vars="true y", var_name="Grades", value_name="Survivability"),
    boxprops=dict(alpha=.7)
)

plt.setp(p.artists, edgecolor = 'k')
plt.setp(p.lines, color='k')



handles, _ = p.get_legend_handles_labels()          # Get the artists.
p.legend(handles, ["Death class (C0)","Survived class (C1)"]) # Associate manually the artists to a label.

sns.move_legend(
    p, "lower center",
    bbox_to_anchor=(.5, 1), ncol=1, title="Seed class", frameon=False,
)

# Rotate x labels to prevent overlapping
plt.xticks(rotation=45)
plt.savefig('../Test_results/Grades/Grades.png', format='png', dpi = 400, bbox_inches = "tight" )
plt.show()
```

## Multi

```python
f_all = []
grades = ['Grade 1', 'Grade 2', 'Grade 3', 'Grade 4', 'Grade 9']

f_all += grades
f_all += [f1, f2, f3]
f_all


y_pred = pd.read_csv("../Test_results/Multi/MLP-5/valid_results_data_frame_4_epo.csv")
x = pd.read_csv("../Test2/Multi/Multi_valid_norm_x.csv")
y_true = np.load('../processed_data_TS_2/y_valid_normalized.npy')

res_xy = x[[f1, f2, f3]]
res_xy['true y'] = y_true
res_xy['score y'] = y_pred['score y']
display(res_xy)


from matplotlib.pyplot import figure
custom_palette = ["#fdcdac", "#b3e2cd"] 

figure(figsize=(8, 6), dpi=400)
plt.rcParams.update({'font.size': 22})

p = sns.boxplot(
    x = "true y",
    y = "score y",
    showmeans=True,
    showfliers=False,
    palette=custom_palette,
    data=res_xy[['true y', 'score y']]
)
p.set_xticklabels(["Death class (C0)","Survived class (C1)"])
p.set(ylabel='Survivability', xlabel='Seed class')
plt.ylim(-0.05,1)


#plt.savefig('../Test_results/Multi/Multi.png', format='png', dpi = 400, bbox_inches = "tight" )

y0 = res_xy[(res_xy['true y'] == 0) & (res_xy['score y'] < 0.71)].shape[0]
y1 = res_xy[(res_xy['true y'] == 1) & (res_xy['score y'] < 0.71)].shape[0]

print("Generated alert from class 0: ", y0, (y0/res_xy[res_xy['true y'] == 0].shape[0])*100)
print("Generated alert from class 1: ", y1, (y1/res_xy[res_xy['true y'] == 1].shape[0])*100)
```
